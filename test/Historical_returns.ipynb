```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_style('white')
sns.set_context(context='talk')
```

```python
df_hist = pd.read_csv('hist_return.csv')
df_hist['Do Nothing'] = np.ones(df_hist.shape[0])
df_hist.head()
```

```python
plt.figure(figsize=(10,7))
plt.plot(df_hist['Year'].values,df_hist['Amt'].values,  label = 'Keep money in bank account')
plt.plot(df_hist['Year'].values,df_hist['sp500 value'].values, label = 'Invest in SP500 Index')
plt.plot(df_hist['Year'].values,df_hist['Do Nothing'].values, label = 'Do Nothing')
plt.plot(df_hist['Year'].values,df_hist['brk value'].values, label = 'Invest in Berkshire Hathway')
plt.legend(loc=(0.1,0.5),fontsize = 14)
plt.margins(.05)
plt.title("Return on $1",fontsize = 15)
plt.xlabel("Year",fontsize = 15)
plt.ylabel("Value of $1",fontsize = 15)
plt.annotate('$1', xy=(1966,1), xytext=(1966,100))

plt.annotate('$16269.81', xy=(2017,16269.81), xytext=(2017-8,15800))

plt.savefig("ret_all.png")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(df_hist['Year'].values,df_hist['Amt'].values,  label = 'Keep money in bank account')
plt.plot(df_hist['Year'].values,df_hist['sp500 value'].values, label = 'Invest in SP500 Index')
plt.plot(df_hist['Year'].values,df_hist['Do Nothing'].values, label = 'Do Nothing')
plt.legend(loc=(0.1,0.55),fontsize = 14)
plt.margins(.05)
plt.title("Return on $1",fontsize = 14)
plt.xlabel("Year",fontsize = 14)
plt.ylabel("Value of $1",fontsize = 14)
plt.annotate('$1', xy=(1966,1), xytext=(1966,4))
plt.annotate('$1', xy=(2017,1), xytext=(2017-2,3))
plt.annotate('$14.46', xy=(2017,14.46), xytext=(2017-4,17))
plt.annotate('$131.85', xy=(2017,131.85), xytext=(2017-9,128))
plt.savefig("ret_two.png")
plt.show()
```
